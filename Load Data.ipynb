#### Load Data

```python
import mne
import os
import warnings

DATAPATH = "D:\\EEG Data\\5mnd mmn\\"
cnt_filenames = []
EEGs = []
#Events = []

warnings.filterwarnings("ignore") # Suppresing warnings: For some reason there are warnings while loading cnt
for i, _filename in enumerate(os.listdir(DATAPATH)):
    if(_filename.endswith(".cnt")):
        cnt_filenames.append(_filename)
        EEGs.append(mne.io.read_raw_cnt(DATAPATH + cnt_filenames[i]))     
        #Events.append(mne.read_events(DATAPATH + cnt_filenames[i]))
        
warnings.filterwarnings("default")
```
